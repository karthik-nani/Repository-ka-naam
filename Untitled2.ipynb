```
#accept 3 subjects marks of the student 
#if he gets>30 in all subjects-->average
#          <30 in any of the subjects-->fail
#avg>75 -->'a'
#avg>65 and avg<75 -->'b'
#avg>50 and avg<65 -->'c'
#avg>30 and avg<50 -->'d'
#else -->'fail'
x = int(input("enter physics marks"))
y = int(input("enter astro physics marks"))
z = int(input("enter nano physics marks"))
if(x>30 and y>30 and z>30):
  avg = (x+y+z)/3
  if(avg>75):
    print("A grade")
  elif(avg>65 and avg<75):
    print("B grade")
  elif(avg>50 and avg<65):
    print("C grade")
  elif(avg>30 and avg<50):
    print("D grade")
  else:
    print("fail")
```
